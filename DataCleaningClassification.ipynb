```python
import pandas as pd
import numpy as np

# declare the file name to read and file name to export
FILE_IN  = "brain_stroke.csv"
FILE_OUT = "brain_stroke_clean.csv"

df = pd.read_csv(FILE_IN)
df.head(3)
```

```python
# convert age to int 
df["age"] = np.round(df["age"]).astype(int)
```

```python
df.head(3)
```

```python
# convert gender, ever_married and redidence type to 0  and 1 
df["gender"] = df["gender"].map({"Male":1, "Female":0}).astype("Int64")
df["ever_married"]   = df["ever_married"].str.lower().map({"yes":1, "no":0}).astype("Int64")
df["Residence_type"] = df["Residence_type"].str.lower().map({"urban":1, "rural":0}).astype("Int64")
```

```python
df.to_csv(FILE_OUT, index=False)
print("Saved:", FILE_OUT)
```
